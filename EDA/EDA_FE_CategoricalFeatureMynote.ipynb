### count or frequency encodingh

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
#col = none means columns taken as none
                                                     
```

```python
train_set.drop(0,axis=1,inplace=True)
```

```python
columns = [1,3,5,6,7,8,9,13]
train_set = train_set[columns]
```

```python
train_set.head()
```

```python
train_set.columns = ['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
#change column name
```

```python
train_set.rename({'Status':'Status_1'},axis=1)
```

```python
for feature in train_set.columns:
    print(feature,':',len(train_set[feature].unique()),"Labels")
```

```python
train_set['Country'].unique() #we get unique country name
```

```python
len(train_set['Country'].unique()) # no we get total number of unique elements in country
```

```python
for feature in train_set.columns:
    print(feature,':',len(train_set[feature].unique()),"Labels")
```

```python
train_set['Country'].unique() # here we get unique country name
# below we get values in each unique country
```

```python
train_set['Country'].value_counts() # we get the total number of values in each country
```

```python
country_map = train_set['Country'].value_counts().to_dict() # we convert above value to dictionary
```

```python
country_map
```

```python
train_set['Country'].map(country_map)
```

```python

train_set['Country'] = train_set['Country'].map(country_map)
```

## Target Guided Ordinal Encoding


```python
import pandas as pd
df = pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df['Cabin'].value_counts()
```

```python
len(df['Cabin'].unique())
```

```python
df['Cabin'].unique()
```

```python
df['Cabin'].astype(str)
```

```python
df['Cabin'].astype(str).str[0]
```

```python
df['Cabin'].astype(str).str[0].unique()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.groupby(['Cabin'])['Survived'].mean() # groupby mean of survived with respect to cabin
```

```python
df
```

```python
# dict ----- key ----> [T,M,A,G,C,F,B,E,D]
#           value ----> [0,1,2,3,4,5,6,7,8]
#output --->
{'T':0,'M':1,'A':2,'G':3}
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
 key = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
s=0
dict={}
for i in key :
        dict[i]=s
        s+=1
```

```python
dictNew={}
for i,k in enumerate(key): # here index is taken as i value is k
    #print(i)
    #print(k)
    dictNew[k]=i
    #print('========')
```

```python
dict
```

```python
dictNew
```

```python
mean_dict = df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_dict
```

```python
k = {k:i for i,k in enumerate(key,1)} # list comperhenssion
```

```python
df['Cabin'] = df['Cabin'].map(k) # dict value assigned to Cabin values 
# so so when key col its value is printed
```
