```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.target_names
```

```python

iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df['species']= iris.target
iris_df.head()
```

```python
# X and y values
X = iris_df.drop(['species'],axis = 1)
y = iris_df['species']
```

```python
from numpy import size


X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.3,random_state=17)

print(f"The dimensions of X_train is {X_train.shape}")
print(f"The dimensions of X_test is {X_test.shape}")
```

```python
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
```

```python
score = dt_classifier.score(X_train, y_train)
print(" R - squared train",score)


y_pred = dt_classifier.predict(X_test)

mse = mean_squared_error(y_test,y_pred)
print("MSE :",mse)
print("RMSE :",mse**(1/2.0))
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt_classifier,filled=True)
plt.show()
```

```python
dt_classifier = DecisionTreeClassifier(criterion= "entropy",max_depth=5)
dt_classifier.fit(X_train, y_train)

score = dt_classifier.score(X_train, y_train)
print(" R - squared train",score)

score = dt_classifier.score(X_test, y_test)
print(" R - squared test",score)

y_pred = dt_classifier.predict(X_test)

mse = mean_squared_error(y_test,y_pred)
print("MSE :",mse)
print("RMSE :",mse**(1/2.0))
```

```python
from sklearn.ensemble import AdaBoostRegressor
clf = AdaBoostRegressor()

clf.fit(X_train,y_train)
score = clf.score(X_train, y_train)
print(f" Train accuracy for {clf.__class__.__name__} is {score}")
score = clf.score(X_test, y_test)
print(f" Test accuracy for {clf.__class__.__name__} is {score}")
print("============================")
```
