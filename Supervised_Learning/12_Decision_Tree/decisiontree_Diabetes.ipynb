```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
X = data.drop('Outcome',axis = 1)
y = data['Outcome']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
dt_classifier = DecisionTreeClassifier()
#train
dt_classifier.fit(X_train, y_train)

#predict
y_pred = dt_classifier.predict(X_test)

score = dt_classifier.score(X_train, y_train)
print(f"train accuracy = {score}")
score = dt_classifier.score(X_test, y_test)
print(f"test accuracy = {score}")
```

```python
mse = mean_squared_error(y_test,y_pred)
print("MSE :",mse)
print("RMSE :",mse**(1/2.0))
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt_classifier,filled=True)
plt.show()
```

```python
dt_classifier = DecisionTreeClassifier(criterion= "entropy",max_depth=4)
dt_classifier.fit(X_train, y_train)

score = dt_classifier.score(X_train, y_train)
print(" R - squared train",score)

score = dt_classifier.score(X_test, y_test)
print(" R - squared test",score)

y_pred = dt_classifier.predict(X_test)

mse = mean_squared_error(y_test,y_pred)
print("MSE :",mse)
print("RMSE :",mse**(1/2.0))
```

```python
from sklearn import tree
plt.figure(figsize=(25,20))
tree.plot_tree(dt_classifier,filled=True,class_names = True)
plt.show()
```
