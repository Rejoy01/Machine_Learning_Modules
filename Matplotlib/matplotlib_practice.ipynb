```python
# Simple bar chart showing revenues of major US tech companies

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company) # xaxis name will change accordingly with cmp name
plt.xlabel("Major Companies")
plt.ylabel("Revenue",fontsize=15)
plt.title("Revenue Model for Top Companies",fontsize=16)
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
plt.show()
```

```python
xpos=np.arange(len(company))
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
# Draw pie chart to track down home expenses
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

plt.figure(figsize=(10,7))
plt.axis("equal") # automatically fitin to default size
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct="%0.2f%%",shadow=True,explode=[0.1,0,0.2,0,0])
plt.legend(loc="best",fontsize="small")
plt.show()
```

```python
#explode used to seperate each exponents
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

plt.figure(figsize=(10,7))
plt.axis("equal") # automatically fitin to default size
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct="%0.2f%%",shadow=True,explode=[0.1,0,0.2,0,0],startangle=45)#startangle for rotate
plt.legend(loc="best",fontsize="small")
#plt.show() # delete plt.show for saving the file
plt.savefig("piechart_new.pdf")#jpg for image
#plt.savefig is used for all (his,bar)
```
