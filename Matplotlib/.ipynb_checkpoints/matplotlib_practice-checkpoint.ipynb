```python
plt.xlabel('Day',fontsize=12)
plt.ylabel('Temperature')
plt.title("Weather info")
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend(loc='best',shadow=True,fontsize="medium")# whenever label is given  legend wil check for label
plt.grid()
plt.show()
```

```python
# Simple bar chart showing revenues of major US tech companies

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.xticks(xpos,company) # xaxis name will change accordingly with cmp name
plt.xlabel("Major Companies")
plt.ylabel("Revenue",fontsize=15)
plt.title("Revenue Model for Top Companies",fontsize=16)
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
plt.show()
```

```python
xpos=np.arange(len(company))
```

```python
xpos #numpy array so we can move the bars with this array
```

```python
plt.xticks(xpos,company) # xaxis name will change accordingly with cmp name
plt.xlabel("Major Companies")
plt.ylabel("Revenue",fontsize=15)
plt.title("Revenue Model for Top Companies",fontsize=16)
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")#width of bar is reduced
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
plt.show()
```

```python
#histogram
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
#blood_sugar.sort()
plt.hist(blood_sugar) # by default number of bins is set to 10
```

```python
plt.hist(blood_sugar,rwidth=0.8)#rwidth
```

```python
plt.hist(blood_sugar,bins=3,rwidth=0.8)
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.8)
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel("Sugar level")
plt.ylabel("Number of patients")
plt.title("Charts")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.8,label=["Men","Women"],color=['red','black'])
plt.legend()
plt.show()
```
