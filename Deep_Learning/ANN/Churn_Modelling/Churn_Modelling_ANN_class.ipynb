```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# Dataset
dataset = pd.read_csv("/content/drive/MyDrive/LuminarTechnolabs/AI/Deep_Learning/ANN/Churn_Modelling/Churn_Modelling.csv")
X=dataset.iloc[:,3:13]
y=dataset.iloc[:,13]
```

```python
y
```

```python
# creat dummy variables:
geography = pd.get_dummies(X["Geography"],drop_first=True)
gender = pd.get_dummies(X["Gender"],drop_first=True)
```

```python
X=pd.concat([X,geography,gender],axis=1)
```

```python
X.drop(["Geography","Gender"],axis=1,inplace=True)
```

```python
X.head()
```

```python
# Split the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
# ANN
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,ReLU,LeakyReLU,Dropout
```

```python
#Initialising the ANN
classifier = Sequential()
```

```python
# Adding input layer
classifier.add(Dense(units=11,activation="relu"))
```

```python
# creating Hidden Layer 1
classifier.add(Dense(units=6,activation="relu"))
```

```python
# creating Hidden Layer 2
classifier.add(Dense(units=4,activation="relu"))
```

```python
#  creating output layer 
classifier.add(Dense(units=1,activation="sigmoid"))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
7000*0.33
```

```python
model_history = classifier.fit(X_train_scaled,y_train,validation_split=0.33,batch_size=10,epochs=100)
```
