```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


```

```python
# dataset
dataset = pd.read_csv('/content/drive/MyDrive/DL/ANN/Churn_Modelling/Churn_Modelling.csv')
dataset.head()
X= dataset.iloc[:,3:13]
y=dataset.iloc[:,13]
```

```python
y.head()
```

```python
# create dummy variable
geography = pd.get_dummies(X["Geography"],drop_first=True)
gender = pd.get_dummies(X["Gender"],drop_first=True)
```

```python
gender
```

```python
X = pd.concat([X,geography,gender],axis=1)
```

```python
X.drop(['Geography','Gender'],axis=1,inplace=True)
```

```python
X.head()
```

```python
# split the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Feature Scaling 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
# ANN
from tensorflow.keras.models import Sequential
from keras.layers import Dense
```

```python
# Intialising the ANN
classifier = Sequential()
```

```python
# Adding input layer 
classifier.add(Dense(units=11,activation = "relu"))
```

```python
#creating Hidden layer1
classifier.add(Dense(units=6,activation='relu'))
```

```python
# creating Hidden Layer2
classifier.add(Dense(units =4,activation='relu'))
```

```python
# creating output layer
classifier.add(Dense(units=1,activation="sigmoid"))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_history = classifier.fit(X_train_scaled,y_train,validation_split=0.33,batch_size = 10,epochs=20)
```
